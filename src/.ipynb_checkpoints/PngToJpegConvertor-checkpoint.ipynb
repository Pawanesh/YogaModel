```python
import os

from PIL import Image
from shutil import copyfile
```

```python
train_dir = '/home/jupyter/YogaModel/dataset/training'
test_dir = '/home/jupyter/YogaModel/dataset/testing'
```

```python
train_jpeg_dir = '/home/jupyter/YogaModel/dataset/training-jpg'
test_jpeg_dir = '/home/jupyter/YogaModel/dataset/testing-jpg'
```

```python
labels = os.listdir(test_dir)
```

```python
for label in labels:
    label_dir = os.path.join(test_dir, label)
    jpg_label_dir = os.path.join(test_jpeg_dir, label)
    if not os.path.isdir(jpg_label_dir): 
        os.mkdir(jpg_label_dir)
    images = os.listdir(label_dir)
    for image in images:
        lowerImage = image
        if lowerImage.lower().endswith('png'):
            img_png = Image.open(os.path.join(label_dir,image))
            name = lowerImage.split('.')
            name = name[len(name)-2]
            name += '.jpg'
            try:
                img_png.save(os.path.join(jpg_label_dir,name))
            except:
                print('ignoring')
                print(os.path.join(label_dir, image))
        else:
            copyfile(os.path.join(label_dir,image), os.path.join(jpg_label_dir,image))
            print('Copied:')
            print(os.path.join(label_dir, image))
            print(os.path.join(jpg_label_dir, image))
```

```python
labels = os.listdir(train_dir)
for label in labels:
    label_dir = os.path.join(train_dir, label)
    jpg_label_dir = os.path.join(train_jpeg_dir, label)
    if not os.path.isdir(jpg_label_dir): 
        os.mkdir(jpg_label_dir)
    images = os.listdir(label_dir)
    for image in images:
        lowerImage = image
        if lowerImage.lower().endswith('png'):
            img_png = Image.open(os.path.join(label_dir,image))
            name = lowerImage.split('.')
            name = name[len(name)-2]
            name += '.jpg'
            try:
                img_png.save(os.path.join(jpg_label_dir,name))
            except:
                print('ignoring')
                print(os.path.join(label_dir, image))
        else:
            copyfile(os.path.join(label_dir,image), os.path.join(jpg_label_dir,image))
            print('Copied:')
            print(os.path.join(label_dir, image))
            print(os.path.join(jpg_label_dir, image))
```
